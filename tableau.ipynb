```python
from io import BytesIO
from urllib.request import urlopen
from zipfile import ZipFile
import pandas as pd
import glob
from pathlib import Path
import os, shutil
```

```python
monthly_data_prefix = ['202001',
'202002',
'202003',
'202004',
'202005',
'202006',
'202007',
'202008',
'202009',
'202010',
'202011',
'202012']

for dt in monthly_data_prefix:
    zipurl = f"https://s3.amazonaws.com/tripdata/JC-{dt}-citibike-tripdata.csv.zip"
    with urlopen(zipurl) as zipresp:
        with ZipFile(BytesIO(zipresp.read())) as zfile:
            zfile.extractall(f"JC-{dt}-citibike-tripdata")
    path = f"JC-{dt}-citibike-tripdata/"
    moveto = "data/"
    files = os.listdir(path)
    files.sort()
    for f in files:
        src = path+f
        dst = moveto+f
        shutil.move(src,dst)
```

```python
##consolidating the files into one DF

all_files = glob.glob("data/" + "*.csv")
li = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

myframe = pd.concat(li, axis=0, ignore_index=True)
```

```python
myframe['StartDate'] = pd.to_datetime(myframe['starttime']).dt.date
myframe['EndtDate'] = pd.to_datetime(myframe['stoptime']).dt.date
myframe.head()
```
